```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
data = {
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
}
df = pd.DataFrame(data)
plt.scatter(df['Age'], df['Height (cm)'])
plt.xlabel('Age')
plt.ylabel('Height (cm)')
plt.title('Age vs. Height')
plt.show()
plt.scatter(df['Age'], df['Weight (kg)'])
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.title('Age vs. Weight')
plt.show()
plt.scatter(df['Height (cm)'], df['Weight (kg)'])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs. Weight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
data = {
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
}
df = pd.DataFrame(data)
plt.bar(df['Name'], df['Age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
plt.bar(df['Name'], df['Height (cm)'])
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.title('Height Distribution')
plt.show()
plt.bar(df['Name'], df['Weight (kg)'])
plt.xlabel('Name')
plt.ylabel('Weight (kg)')
plt.title('Weight Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
data = {
    'Name': people,
    'Age': age,
    'Height (cm)': height,
    'Weight (kg)': weight
}
df = pd.DataFrame(data)
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
plt.hist(df['Height (cm)'], bins=5, edgecolor='black')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Height Distribution')
plt.show()
plt.hist(df['Weight (kg)'], bins=5, edgecolor='black')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Weight Distribution')
plt.show()
```
